```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow_datasets as tfds

import matplotlib.pyplot as plt

fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

print(train_images.shape) 

# normalize [0;255] -> [0;1]
train_images, test_images = train_images / 255.0, test_images / 255.0

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

def show():
    plt.figure(figsize=(10,10))
    for i in range(16):
        plt.subplot(4,4,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(train_images[i], cmap=plt.cm.binary)
        plt.xlabel(class_names[train_labels[i]])
    plt.show()

show()

n_filters = 32

# build the CNN: 5 convolutional layers, 3 polling layers
model = keras.models.Sequential()
model.add(layers.Conv2D(n_filters, (3,3), strides=(1,1), padding="valid", activation='relu', input_shape=(28,28,1)))
model.add(layers.Conv2D(n_filters, (3,3), strides=(1,1), padding="valid", activation='relu', input_shape=(28,28,1)))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(n_filters, 3, activation='relu'))
model.add(layers.Conv2D(n_filters, 3, activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(n_filters, 3, activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))
print(model.summary())

# using Categorical Crossentropy and Adam
loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
optim = keras.optimizers.Adam(lr=0.001)
metrics = ["accuracy"]

model.compile(optimizer=optim, loss=loss, metrics=metrics)

# training
batch_size = 64

# variables to store acc's, losses
train_acc = []
train_loss = []
test_acc = []
test_loss = []

#train for 5 epoch, record losses and accuracies 
for epoch in range(6):
    # train the network on training data
    train_hist = model.fit(train_images, train_labels, epochs=1, batch_size=batch_size, verbose=2)
    # see how the epoch does on the test data
    test_eval = model.evaluate(test_images,  test_labels, batch_size=batch_size, verbose=2)
    
    #store data for each epoch
    train_loss.append(train_hist.history['loss'])
    train_acc.append(train_hist.history['accuracy'])
    
    test_loss.append(test_eval[0])
    test_acc.append(test_eval[1])
    
```

```python
# plot the stored acc's and losses
plt.plot(train_acc) 
plt.plot(train_loss)
plt.plot(test_acc) 
plt.plot(test_loss)
   
plt.title('model accuracy')
plt.ylabel('accuracy/loss')
plt.xlabel('epoch')
plt.legend(['train_acc', 'train_loss', 'test_acc', 'test_loss'], loc='upper left')

plt.show()
```
